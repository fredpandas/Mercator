```python
import plotly.graph_objects as go

O = [0,0,'O']
B = [-4.4906,48.3889,'Brest']
R = [-43.2093,-22.911,'Rio']

fig = go.Figure(go.Scattergeo())
fig.update_layout(height = 210,
                  width = 210,
                  margin = {'r':10,'t':10,'l':10,'b':10},
                  paper_bgcolor = "black",
                  legend = dict(x = 0.85, y = 0.5, bgcolor = 'rgb(116, 128, 169)'),
                  geo=dict(resolution = 50, bgcolor = 'black',
                           showocean = True, oceancolor = 'LightBlue',
                           showlakes = True, lakecolor = 'rgb(48, 82, 196)',
                           showcountries = True, showcoastlines = True,
                           showrivers = True, rivercolor = "rgb(48, 82, 196)",
                           countrycolor = 'rgb(168, 27, 53)',
                           landcolor = 'rgb(79, 158, 60)',
                           lonaxis = dict(showgrid = True,
                                          gridcolor = 'white',
                                          gridwidth = 0.2,
                                          dtick = 10),
                           lataxis = dict(showgrid = True,
                                          gridcolor = 'white',
                                          gridwidth = 0.2,
                                          dtick = 10),
                           projection = dict(type= 'orthographic',
                                  rotation=dict(lon=10,lat=30,roll=5))
                          ))
fig.show()
```

```python
def f(x):
    return log(tan(x/2 + pi/4))
X = [0 + k/10000*(8*pi - 0) for k in range(10001)]
Lat = []
for x in X:
    y = f(0) + 1/tan(pi/3) * (x - 0)
    phi = 2*atan(exp(y)) - pi/2
    Lat.append(phi)
Lon_degr = [x*180/pi for x in X]
Lat_degr = [y*180/pi for y in Lat]
fig.add_trace(go.Scattergeo(lon=Lon_degr,lat=Lat_degr,mode='lines',
                            line=dict(color='orange'),name='loxodromie de cap 30'))
```

```python
fig.write_image('loxodromie.svg')
```
